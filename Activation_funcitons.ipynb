```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Identity Activation Function f(x) = x
x = np.arange(1,50)
y = x
plt.plot(x,y)
plt.title("Identity Activation Function")
plt.show();
```

```python
# Binary Activation_funcitons
x = np.arange(1,50)
y = []
for i in x:
    if(i>25):
        y.append(1)
    else:
        y.append(0)

plt.scatter(x,y)
plt.title("Binary Activation Function")
plt.show();
    
```

```python
# Bipolar Activation_funcitons
x = np.arange(1,50)
y = []
for i in x:
    if(i>25):
        y.append(1)
    else:
        y.append(-1)

plt.scatter(x,y)
plt.title("Bipolar Activation Function")
plt.show();
    
```

```python
# Binary Sigmoidal
x = np.arange(-10,10)
y = 1/(1+np.exp(-x))
plt.scatter(x,y)
plt.title("Binary Sigmoidal Activation function")
plt.show();
```

```python
# Bipolar Sigmoidal
x = np.arange(-10,10)
y = 1/(1+np.exp(-x))/(1+np.exp(-x))
plt.scatter(x,y)
plt.title("Bipolar Sigmoidal Activation function")
plt.show();
```

```python
# Ramp Activation_funcitons
x = np.linspace(-10,10)
y = []
for i in x:
    if(i>1):
        y.append(1)
    elif(i>=0 and i<=1):
        y.append(i)
    else:
        y.append(0)
plt.scatter(x,y)
plt.title("Ramp Activation Function")
plt.show();
```
